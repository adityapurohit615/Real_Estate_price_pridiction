# importing packages

```python
import numpy as np
import pandas as pd
```

```python
housing = pd.read_csv('Boston_Housing.csv')
```

```python
housing.head()
```

```python
housing.drop('Unnamed: 14',axis = 1, inplace = True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for train_index, test_index in split.split(housing, housing["CHAS"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
strat_train_set
```

# Creating a pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(missing_values = np.nan, strategy = 'mean')),
    ('std', StandardScaler())
])
```

```python
final_tr_data = my_pipeline.fit_transform(strat_train_set)
```

```python
final_tr_data.shape
```

```python
X_train = strat_train_set.drop("MEDV", axis = 1)
y_train = strat_train_set["MEDV"].copy()
```

```python
X_train
```

```python
y_train
```

# fitting a model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
some_data = strat_train_set.iloc[:5]
some_labels = y_train.iloc[:5]
prepared_data = my_pipeline.transform(some_data)
model.predict(prepared_data)
```
